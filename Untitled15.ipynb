```python
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
def embed_text(text):
    input_ids = tokenizer(text, return_tensors = 'pt', truncation=True, padding = True)
    with torch.no_grad():
        embeddings = model(**input_ids).last_hidden_state.mean(dim = 1)
    return embeddings.numpy()
```

```python
text_docs = [

             "product is smartwatch with fitness and health monitoring feature",
             "it has a long battery life and comes in various colors"
]
```

```python
doc_embeddings = [embed_text(doc) for doc in text_docs]
text_embedding_pairs = list(zip(text_docs, doc_embeddings))

vectorstore = FAISS.from_embeddings(text_embedding_pairs, lambda x: embed_text(x))
```

```python
def generate_response(query):
  text_docs =  vectorstore.similarity_search(query, k= 1)
  context =text_docs[0].page_content
  response = f"the relevant information for '{query}' is : '{context} "
  return response

query = 'tell me about product'
response = generate_response(query)
print(response)
```
